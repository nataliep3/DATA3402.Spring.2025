1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
  def __init__(self, max_value):
    self.count=0
    self.max_value=max_value

  def increment(self):
    if self.count < self.max_value:
      self.count += 1
    else:
      raise ValueError('Uh oh! The counter has reached its maximum value!')

  def reset(self):
    self.count=0
```

```python
class Counter:
  def __init__(self, max_value):
    self.__count=0
    self.__max_value=max_value

  def increment(self):
    if self.__count < self.__max_value:
      self.__count += 1
    else:
      raise ValueError('Uh oh! The counter has reached its maximum value!')

  def reset(self):
    self.__count=0

  def get_count(self):
    return self.__count

  def get_max_value(self):
    return self.__max_value

  def is_at_max(self):
    return self.__count == self.__max_value
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
class Shape(ABC):
    def __init__(self, x, y):
        self.__x = x
        self.__y = y

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    @abstractmethod
    def get_points(self):
        pass
    @abstractmethod
    def is_inside(self, x, y):
        pass
    @abstractmethod
    def overlaps(self,other):
        pass

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)
        self.__length = length
        self.__width = width

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)

    def get_points(self):
        points = []
        x, y = self.get_x(), self.get_y()
        length, width = self.get_length(), self.get_width()

        num_points = min(16, 2 * (int(length) + int(width)) + 4)
        points.extend([(x, y), (x + length, y), (x + length, y + width), (x, y + width)])

        if num_points > 4:
            side_points = num_points - 4
            for i in range(1, int(length)):
                points.append((x + i, y))
                if len(points) >= num_points:
                    break
            for i in range(1, int(width)):
                points.append((x + length, y + i))
                if len(points) >= num_points:
                    break
            for i in range(1, int(length)):
                points.append((x + length - i, y + width))
                if len(points) >= num_points:
                    break
            for i in range(1, int(width)):
                points.append((x, y + width - i))
                if len(points) >= num_points:
                    break

        return points[:16]

    def is_inside(self, x, y):

        x1 = self.get_x()
        y1 = self.get_y()
        x2 = x1 + self.get_length()
        y2 = y1 + self.get_width()
        return x1 <= x <= x2 and y1 <= y <= y2

    def overlaps(self, other):

        if isinstance(other, Rectangle):
            x1 = self.get_x()
            y1 = self.get_y()
            x2 = x1 + self.get_length()
            y2 = y1 + self.get_width()

            other_x1 = other.get_x()
            other_y1 = other.get_y()
            other_x2 = other_x1 + other.get_length()
            other_y2 = other_y1 + other.get_width()
            return not (x2 < other_x1 or x1 > other_x2 or y2 < other_y1 or y1 > other_y2)

        elif isinstance(other, Circle):
          closest_x = max(self.get_x(), min(other.get_x(), self.get_x() + self.get_length()))
          closest_y = max(self.get_y(), min(other.get_y(), self.get_y() + self.get_width()))

          distance = math.sqrt((other.get_x() - closest_x)**2 + (other.get_y() - closest_y)**2)

          return distance <= other.get_radius()

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.__radius = radius

    def get_radius(self):
        return self.__radius

    def area(self):
        return math.pi * self.__radius**2

    def perimeter(self):
        return 2 * math.pi * self.__radius

    def circumference(self):
        return self.perimeter()

    def get_points(self):
        points = []
        x, y = self.get_x(), self.get_y()
        radius = self.get_radius()

        for i in range(16):
            angle = 2 * math.pi * i / 16
            px = x + radius * math.cos(angle)
            py = y + radius * math.sin(angle)
            points.append((px, py))

        return points
    def is_inside(self, x, y):
        center_x = self.get_x()
        center_y = self.get_y()
        distance = math.sqrt((x - center_x)**2 + (y - center_y)**2)
        return distance <= self.get_radius()

    def overlaps(self, other):
      if isinstance(other, Circle):
        distance = math.sqrt((self.get_x() - other.get_x())**2 + (self.get_y() - other.get_y())**2)
        return distance <= (self.get_radius() + other.get_radius())

class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        super().__init__(x, y)
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3

    def get_side1(self):
        return self.__side1

    def get_side2(self):
        return self.__side2

    def get_side3(self):
        return self.__side3

    def area(self):
        s = (self.__side1 + self.__side2 + self.__side3) / 2
        return math.sqrt(s * (s - self.__side1) * (s - self.__side2) * (s - self.__side3))

    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3

    def get_points(self):
        points = []
        x, y = self.get_x(), self.get_y()
        side1, side2, side3 = self.get_side1(), self.get_side2(), self.get_side3()

        vertex1 = (x, y)
        vertex2 = (x + side1, y)
        vertex3 = (x, y + side2)

        num_points_per_side = 5

        for i in range(num_points_per_side):
            t = i / (num_points_per_side - 1)
            px = (1 - t) * vertex1[0] + t * vertex2[0]
            py = (1 - t) * vertex1[1] + t * vertex2[1]
            points.append((px, py))

        for i in range(num_points_per_side):
            t = i / (num_points_per_side - 1)
            px = (1 - t) * vertex2[0] + t * vertex3[0]
            py = (1 - t) * vertex2[1] + t * vertex3[1]
            points.append((px, py))

        for i in range(num_points_per_side):
            t = i / (num_points_per_side - 1)
            px = (1 - t) * vertex3[0] + t * vertex1[0]
            py = (1 - t) * vertex3[1] + t * vertex1[1]
            points.append((px, py))

        return points[:16]
    def is_inside(self, x, y):
        x1, y1 = self.get_x(), self.get_y()
        x2, y2 = x1 + self.get_side1(), y1
        x3, y3 = x1, y1 + self.get_side2()

        def area(x1, y1, x2, y2, x3, y3):
            return 0.5 * abs((x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)))

        A = area(x1, y1, x2, y2, x3, y3)
        A1 = area(x, y, x2, y2, x3, y3)
        A2 = area(x1, y1, x, y, x3, y3)
        A3 = area(x1, y1, x2, y2, x, y)

        return abs(A - (A1 + A2 + A3)) < 1e-6
```

10. Copy the `Canvas` class from lecture to in a python file creating a `paint` module. Copy your classes from above into the module and implement paint functions. Implement a `CompoundShape` class. Create a simple drawing demonstrating that all of your classes are working.

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]

    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]

    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)

    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)

    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

12. Implement the ability to load/save raster drawings and demonstate that your method works. One way to implement this ability:

   * Overload `__repr__` functions of all objects to return strings of the python code that would construct the object.
   
   * In the save method of raster drawing class, store the representations into the file.
   * Write a loader function that reads the file and uses `eval` to instantiate the object.

For example:

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```
